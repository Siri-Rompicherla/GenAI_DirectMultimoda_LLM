```python
"""
metrics.py

Sprint Health Agent Metrics Library

This module contains reusable functions used to calculate
Sprint Health KPIs from Jira or Azure DevOps exports.

Author: Sirisha Rompicherla
Project: AI Sprint Health Agent
"""

import pandas as pd


# ============================================================
# VELOCITY
# ============================================================

def calculate_velocity(df):
    """
    Velocity = Sum of Story Points for completed work.

    Returns:
        int/float
    """

    completed = df[df["Status"] == "Done"]

    return completed["Story Points"].sum()


# ============================================================
# COMMITTED STORY POINTS
# ============================================================

def calculate_committed_points(df):
    """
    Total story points committed at sprint start.

    Returns:
        int/float
    """

    return df["Story Points"].sum()


# ============================================================
# PREDICTABILITY
# ============================================================

def calculate_predictability(df):
    """
    Predictability measures how much of the committed work
    was actually completed.

    Formula:
    Completed SP / Committed SP * 100

    Returns:
        float
    """

    committed = calculate_committed_points(df)

    completed = calculate_velocity(df)

    if committed == 0:
        return 0

    return round((completed / committed) * 100, 2)


# ============================================================
# CARRY OVER
# ============================================================

def calculate_carryover(df):
    """
    Story points not completed by sprint end.

    Returns:
        float
    """

    incomplete = df[df["Status"] != "Done"]

    return incomplete["Story Points"].sum()


# ============================================================
# CARRY OVER %
# ============================================================

def calculate_carryover_percentage(df):
    """
    Percentage of committed work carried forward.

    Returns:
        float
    """

    committed = calculate_committed_points(df)

    carryover = calculate_carryover(df)

    if committed == 0:
        return 0

    return round((carryover / committed) * 100, 2)


# ============================================================
# BLOCKED WORK %
# ============================================================

def calculate_blocked_percentage(df):
    """
    Percentage of work currently blocked.

    Returns:
        float
    """

    total_items = len(df)

    blocked_items = len(df[df["Blocked"] == "Yes"])

    if total_items == 0:
        return 0

    return round((blocked_items / total_items) * 100, 2)


# ============================================================
# AVERAGE BLOCKED DAYS
# ============================================================

def calculate_average_blocked_days(df):
    """
    Average blocked duration.

    Returns:
        float
    """

    blocked = df[df["Blocked"] == "Yes"]

    if len(blocked) == 0:
        return 0

    return round(blocked["Blocked Days"].mean(), 2)


# ============================================================
# HIGH PRIORITY RISK ITEMS
# ============================================================

def count_high_priority_open_items(df):
    """
    Counts High/Critical items that are not completed.

    Returns:
        int
    """

    risk_items = df[
        (df["Priority"].isin(["High", "Critical"])) &
        (df["Status"] != "Done")
    ]

    return len(risk_items)


# ============================================================
# SPRINT HEALTH SCORE
# ============================================================

def calculate_health_score(df):
    """
    Health score from 0-100.

    This is intentionally simple for MVP.

    Components:
    - Predictability
    - Blocked Work
    - Carry Over

    Returns:
        float
    """

    predictability = calculate_predictability(df)

    blocked_pct = calculate_blocked_percentage(df)

    carryover_pct = calculate_carryover_percentage(df)

    score = (
        (predictability * 0.6)
        + ((100 - blocked_pct) * 0.2)
        + ((100 - carryover_pct) * 0.2)
    )

    return round(score, 2)


# ============================================================
# HEALTH STATUS
# ============================================================

def get_health_status(score):
    """
    Converts score into status.

    Returns:
        string
    """

    if score >= 80:
        return "🟢 Healthy"

    elif score >= 60:
        return "🟡 At Risk"

    return "🔴 Critical"


# ============================================================
# SUMMARY GENERATOR
# ============================================================

def generate_sprint_summary(df):
    """
    Returns sprint KPI summary dictionary.

    Useful for dashboard display and AI prompts.
    """

    score = calculate_health_score(df)

    summary = {

        "Velocity":
            calculate_velocity(df),

        "Committed":
            calculate_committed_points(df),

        "Predictability":
            calculate_predictability(df),

        "Carryover":
            calculate_carryover(df),

        "Carryover %":
            calculate_carryover_percentage(df),

        "Blocked %":
            calculate_blocked_percentage(df),

        "Avg Blocked Days":
            calculate_average_blocked_days(df),

        "High Risk Items":
            count_high_priority_open_items(df),

        "Health Score":
            score,

        "Health Status":
            get_health_status(score)
    }

    return summary
```
